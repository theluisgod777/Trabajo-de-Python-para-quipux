# EJERCICIO BONIFICACIÓN

Puedes pelear y tienes el nivel para pelear contra el dragón, pero debes descansar.

```python
import time
import os

# Función para limpiar la pantalla (opcional)
def limpiar_pantalla():
    os.system('cls' if os.name == 'nt' else 'clear')

# Variables iniciales del jugador
vidas = 3
tiene_espada = True
energia = 60
nivel = 36
pociones = 2
experiencia = 0

def mostrar_vidas():
    """Muestra las vidas como corazones"""
    corazones_llenos = "♥ " * vidas
    corazones_vacios = "♡ " * (3 - vidas)
    return corazones_llenos + corazones_vacios

def mostrar_estado():
    """Muestra el estado actual del jugador"""
    print("\n" + "="*50)
    print("ESTADO DEL JUGADOR")
    print("="*50)
    print(f"Vidas: {mostrar_vidas()}")
    print(f"Espada: {'Sí' if tiene_espada else 'No'}")
    print(f"Energía: {energia}/100")
    print(f"Nivel: {nivel}")
    print(f"Pociones: {pociones}")
    print(f"Experiencia: {experiencia}/100")
    print("="*50 + "\n")

def usar_pocion():
    """Permite usar una poción para recuperar energía"""
    global energia, pociones
    if pociones > 0:
        energia = min(100, energia + 40)
        pociones -= 1
        print(f"¡Has usado una poción! Energía: {energia}/100")
        print(f"Pociones restantes: {pociones}")
    else:
        print("No tienes pociones disponibles.")

def descansar():
    """Permite descansar para recuperar energía"""
    global energia
    print("Descansando...")
    time.sleep(2)
    energia = min(100, energia + 30)
    print(f"¡Te sientes mejor! Energía: {energia}/100")

def entrenar():
    """Permite entrenar para subir de nivel"""
    global nivel, experiencia, energia
    if energia >= 20:
        print("Entrenando...")
        time.sleep(2)
        experiencia += 30
        energia -= 20
        if experiencia >= 100:
            nivel += 1
            experiencia = 0
            print(f"¡Subiste de nivel! Nivel actual: {nivel}")
        else:
            print(f"Ganaste experiencia. Progreso: {experiencia}/100")
    else:
        print("No tienes suficiente energía para entrenar. Descansa primero.")

def buscar_espada():
    """Permite buscar una espada si no la tienes"""
    global tiene_espada, energia
    if not tiene_espada:
        if energia >= 15:
            print("Buscando espada...")
            time.sleep(2)
            tiene_espada = True
            energia -= 15
            print("¡Encontraste una espada!")
        else:
            print("No tienes suficiente energía para buscar. Descansa primero.")
    else:
        print("Ya tienes una espada.")

def verificar_condiciones_dragon():
    """Verifica si puedes pelear contra el dragón"""
    if vidas <= 0:
        print("\nHas perdido todas tus vidas. GAME OVER")
        return False

    if not tiene_espada:
        print("\nNo puedes pelear sin espada. Busca una primero.")
        return False

    if energia >= 30 and energia <= 60:
        print("\nRECOMENDACIÓN: Tu energía está baja. Deberías recoger una poción.")

    if vidas > 0 and tiene_espada and energia >= 70 and nivel > 35:
        print('\n¡Puedes pelear contra el dragón! ¿Estás listo?')
        return True
    elif vidas > 0 and tiene_espada and energia >= 70 and nivel <= 35:
        print('\nPuedes pelear y tienes energía, pero debes subir de nivel antes de pelear contra el dragón.')
        return False
    elif vidas > 0 and tiene_espada and energia < 70 and nivel > 35:
        print('\nPuedes pelear y tienes el nivel para pelear contra el dragón, pero debes descansar.')
        return False
    elif vidas > 0 and tiene_espada and energia < 70 and nivel <= 35:
        print('\nDebes recuperarte y subir de nivel antes de pelear contra el dragón.')
        return False
    else:
        print('\n¡No puedes pelear!')
        return False

def pelear_dragon():
    """Simulación de pelea contra el dragón"""
    global vidas, energia, experiencia
    print("\n" + "="*50)
    print("BATALLA CONTRA EL DRAGÓN")
    print("="*50)
    time.sleep(1)
    print("El dragón lanza una llamarada...")
    time.sleep(2)

    import random
    resultado = random.choice(['victoria', 'derrota', 'empate'])

    if resultado == 'victoria':
        print("¡VICTORIA! ¡Has derrotado al dragón!")
        experiencia += 50
        energia -= 50
        return True
    elif resultado == 'derrota':
        print("El dragón te ha derrotado...")
        vidas -= 1
        energia -= 30
        print(f"Vidas restantes: {mostrar_vidas()}")
        return False
    else:
        print("¡Combate reñido! El dragón huyó...")
        energia -= 40
        experiencia += 20
        return False

def menu_principal():
    """Muestra el menú principal del juego"""
    while vidas > 0:
        mostrar_estado()
        print("MENÚ PRINCIPAL")
        print("="*50)
        print("1. Verificar si puedes pelear contra el dragón")
        print("2. Pelear contra el dragón")
        print("3. Usar poción")
        print("4. Descansar")
        print("5. Entrenar")
        print("6. Buscar espada")
        print("7. Salir del juego")
        print("="*50)

        opcion = input("Elige una opción (1-7): ")

        if opcion == '1':
            verificar_condiciones_dragon()
            input("\nPresiona Enter para continuar...")

        elif opcion == '2':
            if verificar_condiciones_dragon():
                confirmacion = input("\n¿Realmente quieres pelear? (s/n): ").lower()
                if confirmacion == 's':
                    pelear_dragon()
            input("\nPresiona Enter para continuar...")

        elif opcion == '3':
            usar_pocion()
            input("\nPresiona Enter para continuar...")

        elif opcion == '4':
            descansar()
            input("\nPresiona Enter para continuar...")

        elif opcion == '5':
            entrenar()
            input("\nPresiona Enter para continuar...")

        elif opcion == '6':
            buscar_espada()
            input("\nPresiona Enter para continuar...")

        elif opcion == '7':
            print("\n¡Gracias por jugar! Hasta pronto.")
            break

        else:
            print("\nOpción inválida. Intenta de nuevo.")
            input("\nPresiona Enter para continuar...")

    if vidas <= 0:
        print("\nGAME OVER - Has perdido todas tus vidas")
        print("¡Intenta de nuevo!")

# Iniciar el juego
print("\n" + "="*50)
print("BIENVENIDO AL JUEGO DEL DRAGÓN")
print("="*50)
input("Presiona Enter para comenzar...")

menu_principal()
```
